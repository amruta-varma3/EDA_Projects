## Handling Missing Values

```python
Missing_values = [features for features in df.columns if df[features].isnull().sum()>1]

for features in Missing_values:
    print(features,np.round(df[features].isnull().mean(),4))
```

```python
for feature in Missing_values:
    data = df.copy()
    
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### Here With the relation between the missing values and the dependent variable is clearly visible.So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section

```python
# The id features is not required so let's check how many values it have
print("Id of Houses {}".format(len(df.Id)))
```

## How many features are numerical variables?

```python
nf = [feature for feature in df.columns if df[feature].dtypes != 'O'] #numerical feature

print('Number of numerical variables:', len(nf))

df[nf].head()
```

```python
yf = [feature for feature in nf if 'Yr' in feature or 'Year' in feature] #Year feature

yf
```

```python
for feature in yf:
    print(feature, df[feature].unique())
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
## Here we will compare the difference between All years feature with SalePrice

for feature in yf:
    if feature!='YrSold':
        data=df.copy()
        ## We will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## Numerical variables are usually of 2 type
## 1. Continous variable and Discrete Variables

disc_f =[feature for feature in nf if len(df[feature].unique())<25 and feature not in yf+['Id']]
print("Discrete Variables Count: {}".format(len(disc_f)))
```

```python
disc_f
```

```python
df[disc_f].head()
```

```python
## Lets Find the realtionship between them and Sale Price

for feature in disc_f:
    sns.set_palette("Spectral")
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

#### There is a relationship between variable and SalePrice

## Continuous variable

```python
cf=[feature for feature in nf if feature not in disc_f+yf+['Id']]
print("Continuous feature Count {}".format(len(cf)))
```

```python
for feature in cf:
    data=df.copy()
    data[feature].hist(bins=25)
    sns.set_palette('pastel')
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
# We will be using logarithmic transformation


for feature in cf:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
        
```

## Checking Outliers

```python
for feature in cf:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
