### Finding Trends
a.Correlation

```python
new = df.transpose()
corrMatrix=new.corr()
corrMatrix.loc[:,:] = np.tril(corrMatrix, k=-1)
already_in = set()
result = []
for col in corrMatrix:
    perfect_corr = corrMatrix[col][corrMatrix[col] >= 0.9].index.tolist()
    if perfect_corr and col not in already_in:
        already_in.update(set(perfect_corr))
        perfect_corr.append(col)
        result.append(perfect_corr)
```

Above we set the correlation coefficient score to 0.9.
Now that we plot them to observe price trends considering each group as a separate segment
wherein sectors have high correlation with each other.
